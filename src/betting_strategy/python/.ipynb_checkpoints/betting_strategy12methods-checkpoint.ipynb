```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import pandas.io.data as web
%matplotlib inline
```

```python
all_data= pd.read_csv('../../../data/third_party_preds/baidu-091915.csv', encoding='cp1252')
```

```python
all_data['k1 win'] = all_data['Home Win']- (1- all_data['Home Win'])/(all_data['b365H']-1)
all_data['k1 draw'] =all_data['Home Draw']- (1- all_data['Home Draw'])/(all_data['b365D']-1)
all_data['k1 lost'] =all_data['Home Lost']- (1- all_data['Home Lost'])/(all_data['b365A']-1)
result = pd.get_dummies(all_data, columns=['result'])[['result_1', 'result_2', 'result_3']]
```

```python
#kr = pd.DataFrame((all_data[['k1 win','k1 draw','k1 lost']]).values*result.values, columns=result.columns, index=result.index)
#kr.replace([0.0],[np.nan])
#(all_data[['k1 win','k1 draw','k1 lost','result']])
```

```python
all_data['k2 win'] = ((all_data['b365H']+1)*all_data['Home Win']- 1)/(all_data['b365H']+1-1)
all_data['k2 draw'] =((all_data['b365D']+1)*all_data['Home Draw']- 1)/(all_data['b365D']+1-1)
all_data['k2 lost'] =((all_data['b365A']+1)*all_data['Home Lost']- 1)/(all_data['b365A']+1-1)
all_data[['b365H','b365D','b365A','Home Win','Home Draw','Home Lost','k2 win','k2 draw','k2 lost','result']]
```

```python
all_data['select k2'] = all_data[['k2 win','k2 draw','k2 lost']].max(axis=1)
(all_data['select k2']).head()
```

```python
result2 = (all_data[['b365H','b365D','b365A']]-1)
all_data[['b365H-1','b365D-1','b365A-1']]= result2
result2.head()
```

```python
odds =pd.DataFrame(result2.values*result.values, columns=result2.columns, index=result2.index)
odds =odds.replace([0.0],[-1])
all_data[['b365H-odds','b365D-odds','b365A-odds']]= odds
odds.head()
```

```python
#all_data['Ewin']=all_data['b365H']*all_data['Home Win']+all_data['Home Win']-1
#all_data['Edraw']=all_data['b365D']*all_data['Home Draw']+all_data['Home Draw']-1
#all_data['Elost']=all_data['b365A']*all_data['Home Lost']+all_data['Home Lost']-1
#e =pd.DataFrame(all_data[['Ewin','Edraw','Elost']].values*all_data[['k2 win','k2 draw','k2 lost']].values, columns=odds.columns, index=odds.index)
#all_data[['b365H-e','b365D-e','b365A-e']]= e
#e
```

```python
#Best: For each game, we place a unit bet on the outcome with maximum Di, if that
#Di > 0. This is similar to Positive, but places a bet on only one outcome per game.
#Di=Pi-1/odds
all_data['Di win'] = all_data['Home Win']- 1/all_data['b365H']
all_data['Di draw'] = all_data['Home Draw']- 1/all_data['b365D']
all_data['Di lost'] = all_data['Home Lost']- 1/all_data['b365A']

all_data['select Di'] = all_data[['Di win','Di draw','Di lost']].max(axis=1)
(all_data['select Di']).head()
```

```python
all_data['select p'] = all_data[['Home Win','Home Draw','Home Lost']].max(axis=1)
(all_data['select p']).head()
```

```python
#all_data['k2 win select'] = all_data.apply(lambda x : x['k2 win'] if x['k2 win'] >= x['select k2'] and ((x['select k2'] - x['k2 win']) <= 0.05) else 0 , axis=1)
#all_data['k2 win select'] = np.where((all_data['k2 win'] >= all_data['select k2']) | ((all_data['select k2']-all_data['k2 win'])<=0.051)
#                     , all_data['k2 win'], np.nan)
#all_data['k2 draw select'] = np.where((all_data['k2 draw'] >= all_data['select k2']) | ((all_data['select k2']-all_data['k2 draw'])<=0.051)
#                     , all_data['k2 draw'], np.nan)
#all_data['k2 lost select'] = np.where((all_data['k2 lost'] >= all_data['select k2']) | ((all_data['select k2']-all_data['k2 lost'])<=0.051)
#                     , all_data['k2 lost'], np.nan)
#all_data['k2 win select'] = np.where((all_data['k2 win'] > all_data['select k2'])
#                     , all_data['k2 win'], np.nan)
#all_data['k2 draw select'] = np.where((all_data['k2 draw'] > all_data['select k2'])
#                     , all_data['k2 draw'], np.nan)
#all_data['k2 lost select'] = np.where((all_data['k2 lost'] > all_data['select k2'])
#                     , all_data['k2 lost'], np.nan)
#all_data['win select'] = np.where((all_data['k1 win'] > 0) , all_data['k1 win'], np.nan)
#all_data['draw select'] = np.where((all_data['k1 draw'] > 0), all_data['k1 draw'], np.nan)
#all_data['lost select'] = np.where((all_data['k1 lost'] > 0), all_data['k1 lost'], np.nan)
all_data['win select'] = np.where((all_data['Home Win']*all_data['b365H'] >= 1.1) & ((all_data['select p'])>0)
                    , 1+all_data['k1 win'], np.nan)
all_data['draw select'] = np.where((all_data['Home Draw']*all_data['b365D'] >= 1.1) & ((all_data['select p'])>0)
                     , 1+all_data['k1 draw'], np.nan)
all_data['lost select'] = np.where((all_data['Home Lost']*all_data['b365A'] >= 1.1) & ((all_data['select p'])>0)
                     , 1+all_data['k1 lost'], np.nan)
```

```python
k2_res =all_data[['win select','draw select', 'lost select']]
k2_res =k2_res.fillna(0)
k2_res.head()
```

```python
odds.head(10)
```

```python
act_res =pd.DataFrame(odds.values*k2_res.values, columns=odds.columns, index=odds.index)
all_data[['b365H-act_res','b365D-act_res','b365A-act_res']]= act_res
act_res.head()
```

```python
all_data['total'] = act_res['b365H']+act_res['b365D']+act_res['b365A']
```
